```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup
import time

session = HTMLSession()
anime_links = set()  

for x in range(1, 142):
    url = f"https://animeslayerweb.com/page/{x}/?s"
    r = session.get(url)
    print(f"Page {x}: Status Code -> {r.status_code}")
    
    if r.status_code == 200:
        soup = BeautifulSoup(r.content, 'html.parser')
        animelist = soup.find_all('article', class_='bs')
        
        for i in animelist:
          
            a_tag = i.find('a', class_='tip')
            
            if a_tag and a_tag.has_attr('href'):
                link = a_tag['href'].strip()
                anime_links.add(link)  
        
        print(f"تم جمع {len(anime_links)} رابط حتى الآن")
        time.sleep(1)  
```

```python
import pandas as pd
anime_links_df=pd.DataFrame(anime_links)
anime_links_df.to_csv('anime_links.csv')
```

```python

import time

animeslayerdata = []
faildlinks = []
failcount = 0
count = 0


for li in anime_links:
    try:
        s = HTMLSession()
        r = s.get(li)
        print(r.status_code)
        time.sleep(1)
        soup = BeautifulSoup(r.content, 'html.parser')
        print('تم تحميل الصفحة بنجاح')

        # استخراج الأنواع (Genres)
        classes = soup.select('.genxed a')
        itemclasses = [cl.text.strip() for cl in classes] if classes else ["unknown"]

        # استخراج البيانات مع التحقق من وجود العناصر
        title = soup.find('h1').text.strip() if soup.find('h1') else "unknown"

        eps_count = soup.select_one('.spe > span:nth-child(7)').text.strip() if soup.select_one('.spe > span:nth-child(7)') else "unknown"

        publish_date = soup.select_one('span.split:nth-child(11) time').text.strip() if soup.select_one('span.split:nth-child(11) time') else "unknown"

        status = soup.select_one('.spe > span:nth-child(1)').text.strip() if soup.select_one('.spe > span:nth-child(1)') else "unknown"

        season = soup.select_one('.spe > span:nth-child(5) a').text.strip() if soup.select_one('.spe > span:nth-child(5) a') else "unknown"

        studio = soup.select_one('div.spe a').text.strip() if soup.select_one('div.spe a') else "unknown"

        censorship = soup.select_one('.spe > span:nth-child(8)').text.strip() if soup.select_one('.spe > span:nth-child(8)') else "unknown"

        rate = soup.select_one('div.rating strong').text.strip() if soup.select_one('div.rating strong') else "unknown"

        director = soup.select_one('span.split:nth-child(9) a').text.strip() if soup.select_one('span.split:nth-child(9) a') else "unknown"

        eps_period = soup.select_one('.spe > span:nth-child(4)').text.strip() if soup.select_one('.spe > span:nth-child(4)') else "unknown"

        anime_type = soup.select_one('.spe > span:nth-child(6)').text.strip() if soup.select_one('.spe > span:nth-child(6)') else "unknown"

        # إنشاء القاموس
        animedata = {
            'title': title,
            'eps_count': eps_count,
            'publish_date': publish_date,
            'status': status,
            'season': season,
            'studio': studio,
            'Censorship': censorship,
            'rate': rate,
            'director': director,
            'eps_period': eps_period,
            'type': anime_type,
            'classification': itemclasses
        }

        animeslayerdata.append(animedata)
        print('تم إضافة البيانات بنجاح')
        count += 1
        print(f'تم استخراج الأنمي رقم {count}')

    except Exception as e:
        print(f'فشل في معالجة الرابط: {li} - السبب: {e}')
        failcount += 1
        faildlinks.append(li)
        if failcount>=22 :
            break
```

```python
import pandas as pd
animeslayerdf=pd.DataFrame(animeslayerdata)
animeslayerdf.to_csv('Anime_slayer.csv')
```
